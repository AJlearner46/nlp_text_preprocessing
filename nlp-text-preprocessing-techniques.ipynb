3. REMOVING URL

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

4. REMOVE PUNCTUATION

```python
import string
string.punctuation
```

```python
exclude = string.punctuation
```

```python
## slow code
def remove_pun(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

7. Removing stop words

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    s = new_text[:]
    new_text.clear()
    return " ".join(s)
```

8. handlinlg emojis

```python
#remove emojis
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("This dog 😂")
```

10. STEMMING 

```python
#inflection > tense, case, voice, person, number, gender etc.
#converted inflection word to it's root form is called stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return  " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
#convert inflection word to its root word but that root word is word from that language
#in stemming root word may or may not part of root word
# slower than stemmer
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sent = "he was running and eating at same time. he has bad habit of swimmming after playing sfter playing long hours in the sun."
punctuations = "?:!.,;"
sentence_words = nltk.word_tokenize(sent)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)
        
sentence_words
print("{0:20}{1:20}".format("word","lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```
